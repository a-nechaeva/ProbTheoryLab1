#✨***Задание 1, пункт 2***
***
***Условие***.

В городе живут $n+1$ людей. Человек, условный "прародитель", пишет два письма случайно выбранным адресатам, которые образуют "первое" поколение. Те делают то же самое, в результате чего образуется "второе поколение". Если человек, входящий в некоторое "поколение" получает $k$ писем, то он посылает $2k$ писем случайно выбранным адресатам. Найти вероятность того, что прародитель не входит ни в одно из поколений с номерами $1, \dots , r$.

***Upd:*** фразу "Если человек, входящий в некоторое "поколение", получает $k$ писем, то он посылает $2k$ писем случайно выбранным адресатам" формулируется: *на каждое полученное письмо гаражанин готовит два ответа и отправляет двум случайным жителям*.

***
***Решение***



*   Рассмотрим, сколько вариантов отправить письмо из первой пары писем у нас есть. Так как всего жителей у нас $n + 1$, себе письма мы не пишем, значит $n$. Вариантов отправить второе письмо из этой пары у нас $n - 1$, будем считать, что пара писем отправляется двум различным адресатам. Всего вариантов для одной пары писем $n \cdot (n - 1)$.
*   Теперь определим, сколькими способами мы можем отправить пару писем так, чтобы ни одно из них не попало к "прародителю", получим $(n - 1) \cdot (n - 2) $.
*   Вероятность, что ни одно письмо из отправленной пары не попадет к "прародителю":
$$p = \frac{(n - 1) \cdot (n - 2)}{n \cdot (n - 1)} = \frac{n - 2}{n}  $$

*   Вероятность того, что ни одно письмо из $k$ пар не попадет к "прародителю":

$$p^k = \left( \frac{n - 2}{n} \right)^k $$

*   Каждое поколение отправляет в $2$ раза больше писем, чем предыдущее: первое поколение отправило $4$ письма, второе $8$ и так далее, $(r-1)$-ое поколение отправит $2^{r-1}$ пару писем.

*   Вероятность того, что "прародитель" не войдет ни в одно из поколений $1, \dots , r$ запишем по правилу перемножения вероятностей независимых событий:

$$p_{final} = \prod \limits_{k=1}^{r-1} \left( \frac{n - 2}{n} \right)^{2^k} = \left( \frac{n - 2}{n} \right)^{\sum \limits_{k=1}^{r-1} 2^k} = \left( \frac{n - 2}{n} \right)^{2^r - 2}$$

***
***Ответ:***

$$p_{final} = \left( \frac{n - 2}{n} \right)^{2^r - 2}$$

#✨***Задание 2, пункт 4***
***
**Условие**

$n$ супружеских пар случайно садятся за круглый стол. С какой вероятностью их можно разместить так, чтобы мужчины и женщины чередовались, но супруги не сидели рядом?
***
**Решение**



*   Пусть $1, \dots , n$ $-$ множество пар;
*   Всего $(2n)!$ вариантов рассадить $2n$ человек за стол;
*   Требуется, чтобы мужчины и женщины чередовались, рассадим сначала женщин через 1 место друг от друга;
*   Всего $2\cdot n!$ способов рассадить женщин таким образом;
*   Теперь для каждого случая рассадки женщин вычислим число подходящих под условие рассадок мужчин (обозначим $M$);
*   Вероятность удачной рассадки:
$$S = \frac{2 \cdot n! \cdot M}{(2n)!}$$
*   Так как для каждой рассадки женщин $M$ одно, задача симметрична по парам;
*   Пусть есть рассадка женщин: $-1-2-3-4-\dots - n-$;
*   Пусть есть рассадка $-1M_12M_23M_3\dots nM_n-$, где $M_i$ $-$ мужчина принимает значение ${1 \dots n}$, где $M_i=i$ $-$ мужчина из $i$-ой пары;
*   По условию нужно, чтобы супруги не сидели вместе, следовательно, $M_i \neq i, i+1$;
*   Назовем условием $P_i$ ситуацию, когда $M_i = i$ или $M_i = i+1$;
*   Нам нужно вычислить количество ситуаций, в которых не выполнено ни одно из условий $P_i$;
*   $M = n! - L$, где $n!$ $-$ количество случайных рассадок мужчин, а $L$ $-$ количество рассадок, в которых есть хоть одно выполненное условие $P_i$;
*   Посчитаем $L$: мы можем вычислить число рассадок $k$ мужчин и $k$ их жен;
*   Есть $2n$ вариантов посадить какую либо пару супругов вместе, когда женщины уже рассажены ($n$ мужчин и по 2 места);
*   Для каждого из $2n$ случаев есть $(n - 1)!$ вариантов посадить остальных мужчин;
*   Заметим, что эти $2n$ случаев по $(n - 1)!$ рассадок имеют попарное пересечение по всем рассадкам, в которых выполняются 2 условия или более;
*   Мы попадаем в условие задачи ***включения и исключения***;
*   $L$ $=$ количество рассадок с 1 и более условиями $P$ $-$ количество рассадок с 2 и более условиями $P$ $+$ количество рассадок с 3 и более условиями $P$ $- \dots$ и так до $n$ условий $P$;
*   Запишем такую соответствующую формулу:
$L= K_1 - K_2 + K_3 - \dots + (-1)^{k+1}K_k + \dots + (-1)^{n + 1} K_n$,
$K_k$ равно количеству вариантов рассадок $k$ случайных мужчин и их жен, умноженное на $(n - k)!$ $-$ количество вариантов рассадить остальных;
*   $K_1 = 2n \cdot (n - 1)!$;
*   Считаем $K_k$: первый посаженный нами мужчина либо до, либо после своей жены, остальные $k-1$ должны сесть рядом со своими женами;
*   Занумеруем свободные места от жены первого мужчины по часовой стрелке, то есть первый мужчина либо на первом, либо на $n$-ом месте;
*   Рассмотрим оставшиеся варианты сесть на места оставшихся мужчин рядом со своими женами: с собеих сторон у женщины есть место и у него есть номер, на каждый номер можно сесть от двух женщин;
*   Варианты, если первый на месте $1$:
$$2-3|3-4|4-5|\dots -(n-1)|(n-1)-n|n-$$
если первый на месте $n$:
$$-2|2-3|3-\dots -(n-1)|(n-1)-n$$
всего $2n - 3$ варианта и они записаны так, что в силу того, что одному мужу соответствует одна жена и на одно место садится один человек, нельзя выбирать соседние варианты;

*   Следовательно, нас интересует число вариантов выбрать из $2n - 3$ позиции $k-1$ позицию так, чтобы никакие две находящиеся рядом позиции не были выбраны;
*   Такие число равно $C^{k-1}_{2n - k - 1}$;
*   Таким образом,
$$K_k = \frac{2n}{k} C^{k-1}_{2n - k - 1} \cdot (n - k)!$$
делим на $k$, так как первого можно рассмотреть любого из $k$, $(n - k)!$ $-$ количество вариантов рассадки остальных мужчин случайно;
*   В результате получаем такую формулу для вероятности удачной рассадки:
$$S = \frac{2\cdot n! \cdot (n! + (-1)^k \cdot K_k)}{(2n)!}$$,
где $k \in \{1, \dots , n\}$

В квадрат наудачу брошены точки $A, \, B$. Найти вероятность того, что круг, диаметром которого является отрезок $AB$, целиком содержится в квадрате.

***
**Решение**

*   Заметим, что окружность, диаметром которой является отрезок $AB$, можно описать некоторым квадратом, стороны которого будут параллельны сторонам исходного квадрата;
*   Чтобы круг содержался в исходном квадрате, необходимо и достаточно, чтобы квадрат, в который вписана окружность, не выходил за пределы исходного;
*   Достаточно обеспечить то, что два угла описанного квадрата, лежащие на одной диагонали, не выходили за пределы исходного квадрата;
*   Воспользуемся декартовой системой координат: пусть левый нижний угол исходного квадрата располагается в точке $(0; 0)$, а правый верхний в $(1; 1)$ $-$ мы задали квадрат размером $1 \times 1$;

*   Выразим теперь диаметр окружности (расстояние $AB$), этому же значению будет равна сторона описанного вокруг окружности квадрата:
$$r = \sqrt{(x-a)^2 + (y-b)^2}$$

$$\left( \frac{a + x}{2} ; \, \, \, \frac{b + y}{2} \right)$$
*   Координаты правого верхнего угла описанного квадрата:
$$C = (c_1, \, \, c_2) = \left( \frac{a + x}{2} + \frac{r}{2} ; \, \, \, \frac{b + y}{2} + \frac{r}{2} \right)$$
*   И левого нижнего:
$$D = (d_1, \, \, d_2) = \left( \frac{a + x}{2} - \frac{r}{2} ; \, \, \, \frac{b + y}{2} - \frac{r}{2} \right)$$
*   Теперь запишем условия того, что окружность будет содержаться в исходном квадрате:
$$\\
\cases{c_1 < 1\\

$$\cases{\frac{a + x}{2} + \frac{\sqrt{(x-a)^2 + (y-b)^2}}{2} < 1\\

\frac{b + y}{2} > \frac{\sqrt{(x-a)^2 + (y-b)^2}}{2} }$$

*   Возведем обе части неравенств в квадрат (легко убедиться, что это будет равносильные преобразования):

$$\cases{\frac{(x-a)^2 + (y-b)^2}{4} < 1 - a - x + \frac{(a + x)^2}{4} \\

  \frac{(x-a)^2 + (y-b)^2}{4} < 1 - b - y + \frac{(b + y)^2}{4} \\

\frac{(a + x)^2}{4}  > \frac{(x-a)^2 + (y-b)^2}{4} \\

\frac{(b + y)^2}{4} > \frac{(x-a)^2 + (y-b)^2}{4} }$$

*   Продолжим преобразовывать выражения:

  \frac{(x-a)^2}{4} < 1 - b - y + by \\

*   Запишем окончательные неравества:

***
Решение данной задачи было посчитано численно, код представлен ниже. При $n$ $-$ разбиении квадрата на $n^2$ точек, $-$ около $100-120$ получаем вероятность $52 %$ того, что точки будут образовывать круг, лежащий внутри исходного квадрата.
***
**Ответ:**
$$P = 0.52$$

```python
import numpy as np

x_leq = lambda x, y, a, b: x < 1 - ((b - y) ** 2) / (4 * (1 - a))

y_leq = lambda x, y, a, b: y < 1 - ((a - x) ** 2) / (4 * (1 - b))

x_geq = lambda x, y, a, b: x > ((b - y) ** 2) / (4 * a)

y_geq = lambda x, y, a, b: y > ((a - x) ** 2) / (4 * b)


hit = lambda x, y, a, b: x_leq(x, y, a, b) and y_leq(x, y, a, b) and x_geq(x, y, a, b) and y_geq(x, y, a, b)


def big_boss():
    print("enter n:")
    n = int(input())
    a = np.linspace(0, 1, n)
    b = np.linspace(0, 1, n)
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)

    sum = 0

    for i_a in range(1, n - 1):
        for j_b in range(1, n - 1):
            cur_sum = 0
            for i in range(n):
                for j in range(n):
                    if hit(x[i], y[j], a[i_a], b[j_b]):
                        cur_sum += 1
            sum += cur_sum

    print("sum of area = ", sum, "\n")

    print("total sum = ", n ** 2, "\n")

    print("percent = ", sum / ((n ** 2) * ((n - 2) ** 2)) * 100, "%")



big_boss()
```

#✨***Задание 4, пункт 2***

**Условие**

Введем события $A_i = \{ X = i \}, \, B_i = \{ Y= i \}, \, i \geq 0 $. Известно, что для любых $i \geq 0$ и $j \geq 0$ события $A_i$ и $B_j$ -- независимы, при этом

$$P(X=i) = e^{-λ}{ \frac{λ^i}{i!}}, \, λ > 0, \, i \geq 0,$$

$$P(Y=j) = e^{-μ} { \frac{μ^j}{j!}}, \, μ > 0, \, j \geq 0.$$

**Найти** $P(X=i | X + Y = j)$.

***
**Решение**

*   Будем считать, что разбиение дискретно, то есть $i$ и $j$ $-$ целые числа;
*   Запишем теорему Байеса:
$$P(A | B) = \frac{P(B | A) \cdot P(A)}{ P(B)}$$
*   Воспользуемся теоремой Байеса для нашего случая:
$$P = P(X = i | X + Y = j) = \frac{P( X + Y = j| X = i) \cdot P(X = i)}{ P(X + Y = j)}$$

*   Упростим условие, стоящее в числители дроби, получим:
$$P = \frac{P( Y = j - i) \cdot P(X = i)}{ P(X + Y = j)}$$

*   Раскроем знаменатель:
$$P = \frac{P( Y = j - i) \cdot P(X = i)}{\sum \limits_{k=0}^j P(X=k) \cdot P(Y = j - k)}$$

*   Запишем результат в виде:

$$P = \frac{ λ^i \cdot μ^{j-i}}{i! \cdot (j - i)! \cdot \sum \limits_{k=0}^j \frac{λ^k \cdot μ^{j - k}}{k! \cdot (j - k)!} }$$

***
***Ответ:***

$$P = \frac{ λ^i \cdot μ^{j-i}}{i! \cdot (j - i)! \cdot \sum \limits_{k=0}^j \frac{λ^k \cdot μ^{j - k}}{k! \cdot (j - k)!} }$$

**Условие**

Рассмотрите схемы Бернулли при $n \in \{100, 1000, 10000\}$ и $p \in \{0.001, 0.01, 0.1, 0.25, 0.5 \}$. Для каждой пары $(n, p)$ рассчитайте вероятности $P(S_n \in [\frac{n}{2} - \sqrt{npq}, \frac{n}{2} + \sqrt{npq}])$, $P(S_n \leq 5)$ и $P(S_n = k_*)$ по точным формулам (если это технически невозможно, то можно вывести сообщение об этом), где $S_n$ $-$ количество успехов в $n$ испытаниях, $k_*$ $-$ наиболее вероятное количество успехов. Кроме того, каждую вероятность вычислите приближенно с помощью теоремы Пуассона, локальной и интегральной теоремы Муавра-Лапласа. Сравнить результаты, сделать выводы.
***
**Решение**

Запишем неоходимые нам теоремы и формулы:



*   ***Схема Бернулли:***
Проводится $n$ опытов, в каждом из которых может произойти определенное событие ("успех") с вероятностью $p$. Задача $-$ найти вероятность получения ровно $k$ успехов в этих $n$ опытах.

$$P_n(k) = C_n^k p^k (1-p)^{n - k}$$

*   ***Теорема Пуассона:***
Пусть есть последовательность серий испытаний *Бернулли*, где $p$
$-$ вероятность "успеха", $\mu_n$ $-$ количество "успехов".

   Тогда если


1.   $\lim \limits_{n \to \infty} np = λ$;
2.   $λ > 0$;
3.   $\lim \limits_{n \to \infty} P(ω : \mu_n (ω) = k) = e^{-λ} \frac{λ^k}{k!}$

Рассмотрим $n$ независимых испытаний, в каждом из которых событие может произойти с одной и той же вероятностью $p$. Тогда при больших $n$ вероятность можно вычислять по приближенным формулам Муавра-Лапласа.

*   ***Локальная теорема Муавра-Лапласа:***
Вероятность того, что событие наступит в точности $k$ раз, можно вычислить по формуле:
$$P(X=k) ≃ \frac{1}{\sqrt{npq}} 𝜙 \left( \frac{k-np}{\sqrt{npq}} \right)$$
Значения функции $𝜙(x) = \frac{1}{\sqrt{2 \pi}} e^{-x^2 / 2}$ берутся из таблиц, в *Python* есть функция $pdf$, в разделе библиотеки $scipy.stats.norm$.

*   ***Интегральная теорема Муавра-Лапласа:***
Вероятность того, что событие наступит от $k_1$ до $k_2$ раз, можно вычислить по формуле:
$$P(k_1 \leq X \leq k_2) ≃ 𝚽 \left( \frac{k_2 - np}{\sqrt{npq}} \right) - 𝚽 \left( \frac{k_1 - np}{\sqrt{npq}} \right)$$

Здесь $𝚽(x) = \frac{1}{\sqrt{2 \pi}} \int \limits_0^x e^{-z^2 / 2} dz$ $-$ нормированная функция Лапласа, значения так же можно найти в таблице, либо в *Python* $cdf$, в разделе библиотеки $scipy.stats.norm$.

```python
import numpy as np
from math import factorial
from scipy.stats import norm
from decimal import *


#  function gets name for each item in result
_add_names_for_result = lambda _inp: {'Bernoulli ' : _inp[0], 'Poisson ' : _inp[1], 'Local Moivre-Laplace ' : _inp[2], 'Integral Moivre-Laplace ' : _inp[3]}


#  for smart print
def _smart_print(res):
  for key, value in res.items():
    print('\u001b[33;1m', key, '\u001b[0m \n')
    for kkey, vvalue in value.items():
      if isinstance(vvalue, str):
        print('\u001b[31;1m Attention! Probability can not be calculated by ', kkey,'! \u001b[0m ')
      else:
        print("\u001b[34;1m {0} \u001b[0m : \u001b[32;1m {1} \u001b[0m".format(kkey,vvalue))
    print('\u001b[36m ------------------------------------------------------------------------------------- \n')



#  binomial coefficients counter
bin_coef = lambda n, k: factorial(n) / (factorial(k) * factorial(n - k))


#  Bernoulli function
def bernoulli(k, p, n):
    try:
      return bin_coef(n, k) * p ** k * (1 - p) ** (n - k)
    except OverflowError:
      return "Can't count Bernoulli function!"


#  Poisson function
def poisson(k, p, n):
    _lambda = p * n
    try:
      return Decimal(_lambda ** k * np.exp(-_lambda)) / Decimal(factorial(k))

    except OverflowError:
      return "Can't count Poisson function!"


#  integral Moivre--Laplace function
def integral_ML(k1, k2, p, n):
    return norm.cdf((k2 - n * p) / np.sqrt(n * p * (1 - p))) - norm.cdf((k1 - n * p) / np.sqrt(n * p * (1 - p)))


#  local Moivre--Laplace function
def local_ML(k, p, n):
    return norm.pdf((k - n * p) / np.sqrt(n * p * (1 - p))) / np.sqrt(n * p * (1 - p))


#  k in some area of n / 2
def k_in_area_counter(n, p):

  magic_sqrt = np.sqrt(n * p * (1 - p))
  k1 = int(n / 2 - magic_sqrt)
  k2 = int(n / 2 + magic_sqrt)

  _bern = 0
  _pois = 0
  _locM = 0

  for k in range(k1, k2 + 1):
    t = poisson(k, p, n)

    if isinstance(t, str):
      _pois = t
      break

    _pois += t

  for k in range(k1, k2 + 1):
    t = bernoulli(k, p, n)

    if isinstance(t, str):
      _bern = t
      break

    _bern += t

  for k in range(k1, k2 + 1):
    _locM += local_ML(k, p, n)

  _intM = integral_ML(k1, k2, p, n)

  res = [_bern, _pois, _locM, _intM]

  return _add_names_for_result(res)


#  Sn leq 5 fun
def k_leq_5_counter(n, p):

  _bern = 0
  _pois = 0
  _locM = 0

  for k in range(6):
    t = poisson(k, p, n)

    if isinstance(t, str):
      _pois = t
      break

    _pois += t

  for k in range(6):
    t = bernoulli(k, p, n)

    if isinstance(t, str):
      _bern = t
      break

    _bern += t

  for k in range(6):
    _locM += local_ML(k, p, n)

  _intM = integral_ML(0, 5, p, n)

  res = [_bern, _pois, _locM, _intM]

  return _add_names_for_result(res)


#  for k max
def k_max_prob(n, p):
  max_k = round(n * p + 1)
  _bern = 0
  _pois = 0
  _locM = 0


  for k in range(max_k):
    t = poisson(k, p, n)

    if isinstance(t, str):
      _pois = t
      break

    _pois += t

  for k in range(max_k):
    t = bernoulli(k, p, n)

    if isinstance(t, str):
      _bern = t
      break

    _bern += t

  for k in range(max_k):
    _locM += local_ML(k, p, n)

  _intM = integral_ML(0, max_k, p, n)

  res = [_bern, _pois, _locM, _intM]

  return _add_names_for_result(res)


def tester():
    n = [100, 1000, 10000]
    p = [0.001, 0.01, 0.1, 0.25, 0.5]

    for i in range(len(n)):
      for j in range(len(p)):
        print('\u001b[35;1m ', i * len(p) + j + 1, ': n = ', n[i], ' p = ', p[j], '\u001b[0m \n')
        leq_5 = k_leq_5_counter(n[i], p[j])

        in_area = k_in_area_counter(n[i], p[j])

        k_max = k_max_prob(n[i], p[j])

        res = {'P(S less or equal 5)' : leq_5, 'P(S in area of n / 2 plus minus sqrt (pnq))' : in_area, 'P(S = k*), k* - must likely number of successes' : k_max}
        _smart_print(res)




tester()

```
